## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
# 2. Programa para pasar de grados a radianes mediante una entrada por teclado

# Importamos la libreria de math para el valor de pi
import math

# Solicitamos la entrada del valor de arco en grados
grados_valor = input ("Introduce el valor numérico del arco en grados: ")

print ("El valor introducido son", grados_valor,"grados")
# Verificamos que el valor introducido por teclado es de tipo string
# print (type (grados_valor))

# Convertimos el valor de grados a tipo float para poder operar
grados_float = float (grados_valor)
# Verificamos que el tipo de dato es float
# print (type (grados_float))

# Realizamos la operación de conversión a radianes
radianes = grados_float * math.pi/180
# Redondeamos el valor del resultado a 3 decimales
radianes_redondeo = round (radianes, 3)

# Imprimimos el valor de resultado en radianes
print ("El valor de", grados_float,"grados en radianes son: ", radianes_redondeo," radianes" )
```

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
# 3. Programa para el cálculo del área de un paralelogramo con input (base x altura)
# Tenemos 4 tipos de paralelogramos: Cuadrados, rombos, rectángulos y romboides.

# Solicitamos el ingreso de los datos por teclado
print ("Para calcular el área de un paralelogramo, introduce los siguientes datos:")
paral_base = input ("Introduce el valor en milímetros de la base del paralelogramo:")
paral_altura = input ("Introduce el valor en milímetros de la altura del paralelogramo")

# Convertimos los valores introducidos de tipo string a tipo float
paral_base_real = int (paral_base)
paral_altura_real = int (paral_altura)

# Cálculo del area del paralelogramo (Area = base x altura)
paral_area = paral_base_real * paral_altura_real

# Imprimimos el valor del área del paralelogramo
print ("El valor del área del paralelogramo es:", paral_area, "milímetros cuadrados")
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
# 4. Programa para el cálculo del área de un paralelogramo con input (base x altura) con decimales en español (decimal ",")
# Tenemos 4 tipos de paralelogramos: Cuadrados, rombos, rectángulos y romboides.

# Solicitamos el ingreso de los datos por teclado
print ("Para calcular el área de un paralelogramo, introduce los siguientes datos: ")
paral_base = input ("Introduce el valor en milímetros de la base del paralelogramo: ")
paral_altura = input ("Introduce el valor en milímetros de la altura del paralelogramo ")

# Convertimos los valores introducidos de tipo string a tipo float, reemplazando previamente "," por "." en los decimales
paral_base_real = float (paral_base.replace (",","."))
paral_altura_real = float (paral_altura.replace (",", "."))

# Cálculo del area del paralelogramo (Area = base x altura)
paral_area = paral_base_real * paral_altura_real

# Imprimimos el valor del área del paralelogramo redondeado a 3 decimales
print ("El valor del área del paralelogramo es:", round (paral_area,3), "milímetros cuadrados")
```
