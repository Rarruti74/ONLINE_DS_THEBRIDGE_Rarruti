## Ejercicio 1
Para este ejercicio vamos a poner en práctica [las funciones built in](https://docs.python.org/3/library/functions.html).

1. Calcula el máximo de la lista: [4, 6, 8, -1]
2. Suma todos los elementos de la lista anterior
3. Redondea este float a 3 dígitos decimales: 63.451256965
4. Valor absoluto de: -74

```python
print ("1. Vamos a cálcular el max de la siguiente lista")
lista = [4, 6, 8, -1]
# Imprimimos la lista
print ("La lista definida es la siguiente: ", lista)
# Imprimir el valor del max de la lista anterior
print ("El valor máximo de la lista es: ", max(lista))

print ("2. Suma todos los elementos de la lista anterior")
# Imprimimos la lista
print ("La lista anterior es la siguiente: ", lista)
# Suma de los elementos de la lista
print ("La suma de los elementos de la lista es: ", sum(lista))

print ("3. Vamos a redondear a tres decimales un número real")
numero_float = 63.451256965
# Imprimimos el número float y su redondeo
print ("El número flotante es el siguiente: ", numero_float)
print ("El redondeo a tres decimales del número anterior es: ", round (numero_float, 3))

```

## Ejercicio 2
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

```python
# Programa para imprimir en pantalla diferentes peticiones sobre una cadena de texto
cadena_texto = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."
print ("Tenemos la siguiente cadena de texto: ")
print (cadena_texto)
print (" ")

# 1. Pásamos el texto a mayúsculas
print ("1. Pasamos el texto a mayúsculas")
print (cadena_texto.upper ())
print (" ")

# 2. Pásamos el texto a minúsculas
print ("2. Pasamos el texto a minúsculas")
print (cadena_texto.lower ())
print (" ")

# 3. Convertimos la primera letra de cada palabra en mayúsculas y el resto a minúsculas
cadena_texto_lower = cadena_texto.lower ()
print (cadena_texto_lower.title ())
print (" ")

# 4. Crea una lista dividiéndo el texto por los espacios
print ("4. Creamos una lista dividiendo la cadena de texto por los espacios")
lista_texto = cadena_texto.split ()
print (lista_texto)
print (" ")

# 5. Sustituye las "," por ";" en la cadena de texto
print ('5. Sustituye las "," por ";" en la cadena de texto')
print (cadena_texto.replace (",", ";"))
print (" ")

# 6. Elimina las "a" minúsculas en la cadena de texto
print ('6. Elimina las "a" minúsculas en la cadena de texto')
print (cadena_texto.replace ("a", ""))

```

## Ejercicio 3
1. Crea una lista con 3 elementos numéricos
2. Añade un cuarto elemento
3. Calcula la suma de todos
4. Elimina el segundo elemento de la lista
5. Añade otro elemento en la posicion 3 de la lista
6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.
7. Ordena la lista de menor a mayor.
8. BONUS: Ordena la lista de mayor a menor.  
(*PISTA: Para hacer 8, tendrás que buscar en la información del método...*)  
  
  
NOTA: Pinta por pantalla el resultado de cada paso.

```python
# 1. Crea una lista con 3 elementos numéricos
print ("1. Creamos una lista con 3 elementos numéricos")
lista_num = [1, 2, 3]
print (lista_num)

# 2. Añade un cuarto elemento a la lista
print ("2. Añadimos un cuarto elemento a la lista", lista_num)
print ("Añadimos el número 4 a la lista: ")
lista_num.append (4)
print (lista_num)

# 3. Cálcula la suma de todos los elementos de la lista
print ("3. Cálcula la suma de todos los elementos de la lista", lista_num)
suma_lista_num = sum (lista_num)
print ("La suma de los elementos de la lista es:", suma_lista_num)

# 4. Elimina el segundo elemento de la lista
print ("4. Elimina el segundo elemento de la lista", lista_num)
lista_num.pop (1)
print ("La nueva lista es: ", lista_num)

# 5. Añade otro elemento en la tercera posición de la lista
print ("5. Añade otro elemento en la tercera posición de la lista", lista_num)
lista_num.insert (2, 5)
print ("Añadimos el número 5 a la lista: ", lista_num)

# 6. Crea otra lista con 4 elementos numéricos y concatenala con la anterior
print ("6. Creamos otra lista con 4 elementos numéricos")
lista_num_2 = [6, 8, 7, 9]
print (lista_num_2)
print ("Concatenamos las dos listas:")
lista = lista_num + lista_num_2
print (lista)

# 7. Ordena la lista resultante de menor a mayor
print ("7. Ordenamos la lista de menor a mayor")
lista.sort ()
print (lista)

# 8. Ordena la lista resultante de mayor a menor
print ("8. Ordenamos la lista de mayor a menor")
lista.reverse ()
print (lista)
```
