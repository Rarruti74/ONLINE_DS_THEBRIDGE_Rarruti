## Ejercicio 1
Eleva todos los elementos de la lista al cuadrado
> ```ejer_2 = [1,2,3,4,5]```

`````python
# 1. Creamos la siguiente lista y la printamos
ejer_2 = [1, 2, 3, 4, 5]
print ("Tenemos la siguiente lista, ejer_2:", ejer_2)
# print (type (ejer_2)) # Confirmamos que se trata de una lista

list = []                    # Creamos la nueva lista vacia "list"
for i in ejer_2:             # Recorremos con un ciclo for, todos los elementos de  la lista "eje_2"
      list.append(i**2)      # Elevar al cuadrado cada elemento y lo añadimos a la nueva lista "list" en cada ciclo

print ("La lista con los elementos al cuadrado es:", list)
`````
